# Your First LLM Call
Send text to Gemini, get text back.

```python
# !pip install google-genai
```

```python
import os
from google import genai

os.environ["GOOGLE_API_KEY"] = "YOUR_KEY_HERE"
client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="What is an IFC file? Answer in 2 sentences.",
)
print(response.text)
```

## System Prompt
Give the LLM a role -- it changes how it answers.

```python
from google.genai import types

response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="What happens if a staircase riser is too high?",
    config=types.GenerateContentConfig(
        system_instruction="You are a building code expert. Max 20 words.",
    ),
)
print(response.text)
```

## Key Takeaways

1. LLM call = text in, text out
2. System prompt = give it a role
3. **Problem:** we get free text, but code needs data
4. **Next:** make the LLM return JSON
